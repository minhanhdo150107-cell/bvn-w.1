```python
chieu_dai=12345
chieu_rong=5789
pi=3.14
r=chieu_rong / 2
S_hcn = chieu_dai * chieu_dai
S_hcn = round(S_hcn,2)
S_tron = pi * r * r
S_tron = round(S_tron,2)
S_con_lai = S_hcn -S_tron
S_con_lai = round(S_con_lai,2)
print("Diện tích hình chữ nhât là :",S_hcn)
print("Diện tích hình tròn là :",S_tron)
print("Diện tích phần còn lại là:",S_con_lai)
```

```python
c = input("Nhập vào một ký tự:")
if   c.isupper():#Nếu c là chữ in hoa
    print("Chữ cái thường:",c.lower())
elif c.islower():#Nếu c là chữ thường
    print("Chữ cái in hoa:",c.upper())

```

```python
c = input("Nhập một ký tự")
if c.isalpha():#nếu c là một chữ cái
    print(f"{c} là ký tự alphabet.")
else :#nếu c không phải chữ cái
    print(f"{c} không phải là ký tự alphabet.")
```
